```python
# 2.write a program to search a word and replace with another word for all the occurences

with open('colour.txt','r') as f :
    filedata=f.read()
print("initial file data\n..................")
with open('colour.txt','w') as f :
    f.write(filedata)
print(filedata)

word=input("\nenter a word you want to replace :\n")
newword=input("\nenter a word you want to insert :\n.............\n")
print("\nnew file data\n...............")
filedata=filedata.replace(word,newword)
with open('colour.txt','w') as f :
    f.write(filedata)
print(filedata)


#OUTPUT

#-------------------------------

# initial file data
# ..................
# RED EYED ANU IS DRINKING RED COLOUR JUICE.
# RED EYED ROM DRIVES A RED CAR
# BLOOD HAVE RED COLOUR
# RED IS MY FAVOURITE COLOUR
# I HAVE A SHALL IN RED COLOUR

# enter a word you want to replace :
# RED

# enter a word you want to insert :
# .............
# BLUE

# new file data
# ...............
# BLUE EYED ANU IS DRINKING BLUE COLOUR JUICE.
# BLUE EYED ROM DRIVES A BLUE CAR
# BLOOD HAVE BLUE COLOUR
# BLUE IS MY FAVOURITE COLOUR
# I HAVE A SHALL IN BLUE COLOUR
```

```python
# 3.write a program to delete coment lines(lines beginning with #) from a file

import re
with open('textmodel.txt','r') as file:
    lines=[]
    print("\ninitial file data :\n............................\n")
    for l in file:
        print(l,end='')
        lines.append(l)
        
with open('textmodel.txt','w') as file:
    for l in lines:
        if l.startswith('#'):
            pass
        else:
            file.write(l)
            
with open('textmodel.txt','r') as file:
    lines=[]
    print("\n\nfinal file data :\n...............................\n")
    for l in file:
        print(l)
        lines.append(l)
        
#output

#.................

# initial file data :
# ............................

# IT IS SO HARD
# #THIS IS A COMMENT LINE
# THIS IS NOT A COMMEND LINE
# #I AM SO HAPPY TO SEE YOU

# final file data :
# ...............................

# IT IS SO HARD

# THIS IS NOT A COMMEND LINE
```

```python
# 4.write a program to count the frequencies of each word from a file
with open('frequency.txt','r')as file:
    lines=file.readlines()
    freq={}
    for l in lines:
        words=l.split()
        for word in words:
            if word in freq:
                freq[word]=freq[word]+1
            else:
                freq[word]=1
    print(freq)
```
